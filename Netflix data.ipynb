```python
#summary of all the columns

def check (df):
    l=[]
    columns=df.columns
    for col in columns:
        dtypes=df[col].dtypes
        nunique= df[col].nunique()
        sum_null=df[col].isnull().sum()
        l.append([col,dtypes,nunique,sum_null])
    df_check=pd.DataFrame(l)
    df_check.columns=["column", "dtypes", "nunique","sum_null"]
    return df_check
check(netflix)
```

```python
#movie type by country
movie_type_by_relyear = pd.crosstab(netflix["type"], netflix["release_year"].nunique())
movie_type_by_relyear
```

```python
# movie type by months added
movie_by_month = pd.crosstab(netflix["type"], netflix["month"])
movie_by_month
```

```python
plt.figure(figsize=(14, 8))
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 
                'August', 'September', 'October', 'November', 'December']
sns.countplot(data= netflix, x='month', hue = 'type', palette='Set1', order=months_order)
plt.title('Count of Movies and TV Shows Added by months')
plt.xlabel('Month added')
plt.ylabel('type')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
for i in netflix.select_dtypes(exclude='number').columns:
    print(netflix[i].value_counts())
    print('-' * 100)
```

```python
# movie type by year added
movie_by_year = netflix.groupby("year_added")["type"].count()
movie_by_topyear = movie_by_year.sort_values(ascending=False)
movie_by_topyear
```

```python
plt.figure(figsize=[10,6])
sns.lineplot(x=movie_by_topyear.index, y=movie_by_topyear.values)
plt.xlabel('Year_added')
plt.ylabel('Count')
plt.title('Movie Count by Year')
plt.show();
```

```python
plt.figure(figsize=[10,6])
sns.barplot(x=movie_by_topyear.index, y=movie_by_topyear.values)
plt.xlabel('Year_added')
plt.ylabel('Count')
plt.title('Movie Count by Year')
plt.show();
```

```python
plt.figure(figsize =(10,6))
sns.countplot(x = "rating", data = netflix, hue = "type", palette='Set2')
plt.xticks(rotation = 45)
plt.title("Movies rating by type", fontdict= {"fontsize": 20})
plt.show()
```

```python
# Calculating the top 5 country based on type of movies

type_country = netflix.groupby("country")["type"].size()
top_5_country = type_country.sort_values(ascending=False).head(5)

# Printing the top 5 cities by sales
print(top_5_country)
```

```python
plt.figure(figsize=[10,6])
sns.barplot(x=top_5_country.index, y=top_5_country.values)
plt.title(" Top 5 Country most watched movie type", fontdict={"fontsize": 20})
plt.show();
```

```python
# Distribution of type
count_data = netflix['type'].value_counts()
total_count = len(netflix)

plt.figure(figsize=(20, 10))
plt.pie(count_data, labels=count_data.index, autopct=lambda p: f'{int(p * total_count / 100)}\n({p:.0f}%)', startangle=40, pctdistance=0.85,
        wedgeprops={'width': 0.4, 'edgecolor': 'w'})

centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Distribution of Movie Type')
plt.axis('equal')
plt.show();
```

```python
label_data = netflix['type'].value_counts()
explode = (0.1, 0.1)
plt.figure(figsize=(20, 10), facecolor='#F5F5F5')
patches, texts, pcts = plt.pie(label_data,
                               labels=label_data.index,
                               pctdistance=0.85,
                               shadow=True,
                               startangle=40,
                               explode=explode,
                               autopct='%1.1f%%',
                               textprops={'fontsize': 10,
                                          'color': 'black',
                                          'weight': 'bold',
                                          'family': 'serif'})
plt.setp(pcts, color='white')
hfont = {'fontname': 'serif', 'weight': 'bold'}
plt.title('Label', size=10, **hfont)
centre_circle = plt.Circle((0, 0), 0.40, fc='#F5F5F5')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```
